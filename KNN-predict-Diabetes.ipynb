# KNN - Predict whether a person will have diabetes or not

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

load data base

```python
dataset = pd.read_csv("diabetes.csv")
print(dataset.shape)
dataset.head()
```

Values of columns like Glucose, Blood pressure cant have zeros as it will affect the outcome
replace such values with the mean oif the respective column

```python
# replace zeros
zeros_not_accepted = dataset.columns[1:6]
for column in zeros_not_accepted:
    dataset[column] = dataset[column].replace(0, np.nan)
    dataset[column].fillna(int(dataset[column].mean(skipna=True)),inplace=True) 
```

```python
dataset.head()
```

split dataset into train and test datas

```python
X = dataset.iloc[:, 0:8]
y = dataset.iloc[:,8]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

Rule of thumb : 
ANy algorithm that computes distance or assumes normality, scale your features
#Feature Scaling

```python
#  standard scalar bring every data to 1 to -1
dataset.head(25)
```

```python
# Feature Scaling 
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

#Defining model : Init K-NN

CHoose 11 as k(n_neighbors) because  sq_rt(len(y_train)) = 12 

and make sure k should not be even so 12-1 = 11

```python
classifier  = KNeighborsClassifier(n_neighbors=11, p=2, metric="euclidean")
```

**Fit Model**

```python
classifier.fit(X_train, y_train)
```

**##### Predict the test results**

```python
y_pred = classifier.predict(X_test)
y_pred
```

Evaluate Model

```python
# in confusion matrix =>  first true test value is passed and then predicted value from the model

cm = confusion_matrix(y_test, y_pred)
print( "cm => ", cm)
f1Score = f1_score(y_test, y_pred)
print("F1_Score ->", f1Score)
```

In the Above CM output,  compare it with 
[[TP  FP] 
[FN  TN]]

Then we  will know that TP = 94, FP = 13, FN = 15, TN = 32

TP and TN is very much Important

```python
accuracy = accuracy_score(y_test, y_pred)
print("accuracy_score = ", accuracy)
```
